```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('cleaned_food_delivery.csv')
```

```python
def age_group(age):
    if age < 25:
        return 'Young Adult'
    elif age < 40:
        return 'Adult'
    elif age < 55:
        return 'Middle Age'
    else:
        return 'Senior'

df['Age_Group'] = df['Customer_Age'].apply(age_group)
```

```python
def delivery_category(time):
    if time <= 60:
        return 'Fast'
    elif time <= 180:
        return 'Average'
    else:
        return 'Slow'

df['Delivery_Category'] = (
    df['Delivery_Time_Min']
    .apply(delivery_category)
)
```

```python
def revenue_category(value):
    if value < 1000:
        return 'Low'
    elif value < 3000:
        return 'Medium'
    else:
        return 'High'

df['Revenue_Category'] = (
    df['Final_Amount']
    .apply(revenue_category)
)
```

```python
def profit_category(margin):
    if margin < 0:
        return 'Loss'
    elif margin < 0.2:
        return 'Low Profit'
    else:
        return 'High Profit'

df['Profit_Category'] = (
    df['Profit_Margin']
    .apply(profit_category)
)
```

```python
df[['Age_Group',
    'Delivery_Category',
    'Revenue_Category',
    'Profit_Category']].head()
```

```python
df.to_csv(
    'final_food_delivery.csv',
    index=False
)
```
