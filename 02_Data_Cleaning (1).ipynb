```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('ONINE_FOOD_DELIVERY_ANALYSIS.csv')

df.shape
```

```python
df_clean = df.copy()
```

```python
df_clean['Order_Date'] = pd.to_datetime(
    df_clean['Order_Date'],
    errors='coerce'
)
```

```python
df_clean['Order_Time'].dropna().head(10)
```

```python
num_cols = [
    'Customer_Age',
    'Delivery_Time_Min',
    'Distance_km',
    'Order_Value',
    'Discount_Applied'
]

for col in num_cols:
    df_clean[col] = df_clean[col].fillna(
        df_clean[col].median()
    )
```

```python
cat_cols = [
    'Customer_Gender',
    'City',
    'Area',
    'Cuisine_Type',
    'Payment_Mode'
]

for col in cat_cols:
    df_clean[col] = df_clean[col].fillna(
        df_clean[col].mode()[0]
    )
```

```python
df_clean['Final_Amount'] = (
    df_clean['Order_Value']
    -
    df_clean['Discount_Applied']
).clip(lower=0)
```

```python
df_clean['Restaurant_Rating'] = (
    df_clean['Restaurant_Rating']
    .clip(upper=5)
)
```

```python
df_clean.isnull().sum().sort_values(ascending=False)
```

```python
df_clean.to_csv(
    'cleaned_food_delivery.csv',
    index=False
)
```
